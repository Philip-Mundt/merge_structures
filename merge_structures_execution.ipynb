```python
# imports
import MDAnalysis as mda
import numpy as np
from MDAnalysis.lib.distances import capped_distance
import nglview as nv
import os
import warnings
warnings.filterwarnings('ignore')

# Import our custom functions
from merge_structures_final import merge_pdb_files
```

```python
# Define input files and protein names
pdb_files = [
    "input_structures/single_comp_structures/hpl2_config_260.15K.pdb",
    "input_structures/single_comp_structures/lin13_config_260.15K.pdb",
    "input_structures/single_comp_structures/lin65_config_260.15K.pdb",
    "input_structures/single_comp_structures/met2_config_260.15K.pdb"
]

protein_names = ["hpl2", "lin13", "lin65", "met2"]

# Define output path
output_path = "output/merged_structure_260.15K.pdb"

print("Starting structure merging process...")
print(f"Processing {len(pdb_files)} PDB files")
```

```python
# Execute the merging process
merged_universe = merge_pdb_files(
    pdb_files=pdb_files,
    protein_names=protein_names,
    output_path=output_path,
    min_distance=5.0  # Minimum separation between aggregates
)
```

```python
# Visualize the merged structure
if merged_universe is not None:
    print("\nDisplaying merged structure...")
    view = nv.show_mdanalysis(merged_universe)
    view
```

```python
# Verify segment IDs in the merged structure
if merged_universe is not None:
    print("\nVerifying segment IDs...")
    unique_segs = np.unique(merged_universe.atoms.segids)
    print(f"Found {len(unique_segs)} unique segment IDs:")
    for seg in sorted(unique_segs):
        count = len(merged_universe.atoms[merged_universe.atoms.segids == seg])
        print(f"  {seg}: {count} atoms")
```

```python
# Check box dimensions
if merged_universe is not None:
    print("\nBox dimensions:")
    print(f"  {merged_universe.atoms.dimensions}")
```

```python
# Display box dimensions after merging
if merged_universe is not None:
    print("\nFinal box dimensions after concentration-preserving scaling:")
    print(f"  {merged_universe.atoms.dimensions}")
```
